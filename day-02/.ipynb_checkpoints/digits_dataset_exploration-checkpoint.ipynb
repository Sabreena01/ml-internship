# Day 2: Exploring the Digits Dataset

## Objective

The objective of this notebook is to explore the handwritten digits dataset provided by Scikit-learn. In this exercise, I learned how images are stored as arrays, how to visualize them using Matplotlib, how to manipulate image arrays using NumPy, analyze pixel values, and compute the average image for each digit.

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
```

## Step 1: Load the Digits Dataset

The first step is to load the built-in handwritten digits dataset from Scikit-learn. This dataset contains images of handwritten digits (0–9) and their corresponding labels.

Each image has a size of **8 × 8 pixels**, where each pixel contains an intensity value ranging from **0 to 16**.

```python
digits = load_digits()
```

```python
print(digits)
```

## Step 2: Explore the Dataset

After loading the dataset, I checked the shape of the data.

This helps me understand:

- Number of images in the dataset
- Size of each image
- Number of labels

The dataset contains **1797 handwritten digit images**.

```python
print("Shape of data:", digits.data.shape)
print("Shape of images:", digits.images.shape)
print("Shape of target:", digits.target.shape)
```

## Step 3: View an Image as a NumPy Array

An image is simply a matrix of numbers.

Each number represents the brightness of one pixel.

A value close to **0** represents a dark pixel, while values close to **16** represent brighter pixels.

```python
print(digits.images[0])
```

## Step 4: Display Sample Images

Next, I displayed multiple handwritten digit images using Matplotlib.

Each image is shown together with its correct label to understand how the dataset looks visually.

```python
plt.figure(figsize=(10, 4))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(digits.images[i], cmap='gray')
    plt.title(f"Label: {digits.target[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
image = digits.images[0]

print(image)
```

```python
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

## Step 5: Crop an Image

Image cropping is performed using NumPy array slicing.

Cropping removes selected rows or columns from an image and helps understand how images can be manipulated using arrays.

```python
cropped = image[2:6, 2:6]

plt.imshow(cropped, cmap='gray')
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

## Step 6: Flip the Image

Using NumPy slicing, I flipped the image in two different ways:

- Horizontal Flip
- Vertical Flip

This demonstrates how image transformations can be performed without using any image editing software.

```python
flipped_horizontal = image[:, ::-1]

plt.imshow(flipped_horizontal, cmap='gray')
plt.title("Horizontally Flipped")
plt.axis('off')
plt.show()
```

```python
flipped_vertical = image[::-1, :]

plt.imshow(flipped_vertical, cmap='gray')
plt.title("Vertically Flipped")
plt.axis('off')
plt.show()
```

## Step 7: Modify Pixel Values

In this step, I changed some pixel values to **0**.

Since pixel value 0 represents black, the modified area becomes darker.

This experiment helped me understand how changing pixel values directly affects the appearance of an image.

```python
modified = image.copy()

modified[2:6, 3:5] = 0

plt.imshow(modified, cmap='gray')
plt.title("Pixels Set to Zero")
plt.axis('off')
plt.show()
```

## Step 8: Plot Histograms of Pixel Values

A histogram shows how frequently different pixel intensity values occur in an image.

This helps understand:

- Distribution of dark pixels
- Distribution of bright pixels
- Overall brightness of the image

Most pixels are close to zero because the background of handwritten digit images is black.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

# Histogram of first image
plt.subplot(1,2,1)
plt.hist(digits.images[0].ravel(), bins=17)
plt.title("Histogram - Image 0")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")

# Histogram of second image
plt.subplot(1,2,2)
plt.hist(digits.images[1].ravel(), bins=17)
plt.title("Histogram - Image 1")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Step 9: Compute the Average Image for Each Digit

Finally, I computed the average image for every digit (0–9).

For each digit:

- All images belonging to that digit were selected.
- The mean pixel value was calculated.
- The resulting average image was displayed.

These average images appear slightly blurry because they represent the average of many different handwriting styles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure with 10 subplots (one for each digit)
fig, axes = plt.subplots(2, 5, figsize=(12, 5))

# Flatten the axes array for easy looping
axes = axes.ravel()

# Loop through digits 0 to 9
for digit in range(10):

    # Select all images of the current digit
    digit_images = digits.images[digits.target == digit]

    # Compute the average image
    average_image = np.mean(digit_images, axis=0)

    # Display the average image
    axes[digit].imshow(average_image, cmap="gray")
    axes[digit].set_title(f"Average {digit}")
    axes[digit].axis("off")

plt.tight_layout()
plt.show()
```
