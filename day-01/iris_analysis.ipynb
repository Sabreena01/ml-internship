```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()

# Convert it into a Pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the flower species names
df["species"] = iris.target_names[iris.target]

# Display the first 5 rows
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["sepal length (cm)"], bins=10)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["sepal length (cm)"], df["petal length (cm)"])

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df["sepal length (cm)"])

plt.title("Box Plot of Sepal Length")
plt.ylabel("Sepal Length (cm)")

plt.show()
```

# Conclusion

- The Iris dataset contains 150 flower samples with four numerical features and one species column.
- Basic statistics were generated using Pandas.
- A histogram showed the distribution of sepal length.
- A scatter plot showed a positive relationship between sepal length and petal length.
- A box plot summarized the distribution of sepal length and showed no significant outliers.
